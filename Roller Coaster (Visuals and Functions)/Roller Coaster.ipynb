4. Write a function that will plot the ranking of the top `n` ranked roller coasters over time as lines. Your function should take a number `n` and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   For example, if `n == 5`, your function should plot a line for each roller coaster that has a rank of `5` or lower.
   
   Call your function with a value of `n` and either the wood ranking or steel ranking DataFrame.

```python
# 4
years = [2013,2014,2015,2016,2017,2018]

ranks = [ranking_wood[ranking_wood['Year of Rank'] == year]['Points'].sort_values(ascending = False)[:5] for year in years]


# Create a function to plot top n rankings over time
def plot_n_best_rankings(df, n = 5):
    years = [2013,2014,2015,2016,2017,2018]
    ranks = [df[df['Year of Rank'] == year]['Points'].sort_values(ascending = False).reset_index(drop = True)[:n]\
    for year in years]
    legend = []
    for i in range(n):
        series = [rank[i] for rank in ranks]
        plt.plot(years, series)
        legend.append(f'Rank {i+1}')
    plt.legend(legend)
    plt.show()





# Create a plot of top n rankings over time
plot_n_best_rankings(ranking_wood,10)
```

9. `.scatter()` is another useful function in matplotlib that you might not have seen before. `.scatter()` produces a scatter plot, which is similar to `.plot()` in that it plots points on a figure. `.scatter()`, however, does not connect the points with a line. This allows you to analyze the relationship between two variables. Find [`.scatter()`'s documentation here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

   Write a function that creates a scatter plot of two numeric columns of the roller coaster DataFrame. Your function should take the roller coaster DataFrame and two-column names as arguments. Make sure to include informative labels that describe your visualization.
   
   Call your function with the roller coaster DataFrame and two-column names.

```python
# 9
# Create a function to plot scatter of any two columns
def plot_scatter(dim_1, dim_2,zoom=None):
    plt.scatter(roller_coasters[dim_1], roller_coasters[dim_2])
    plt.xlabel(dim_1)
    plt.ylabel(dim_2)
    plt.title(f'Sample roller coasters\' {dim_1} vs. {dim_2}')
    #adding the zoom option to avoid seeing the data panel from too high if outliers (like there is in the 'height' series)
    if zoom:
        plt.axis(zoom)
    plt.show()

# Create a scatter plot of roller coaster height by speed (see difference between zoomed and unzoomed view)
plot_scatter('speed', 'height')
plot_scatter('speed', 'height',[0,250,0,200])
```

What roller coaster seating type is most popular? And do different seating types result in higher/faster/longer roller coasters?

```python
# Most popular seating type: 'Sit Down', by far
plt.barh(y = roller_coasters['seating_type'].value_counts().sort_values().index,\
         width = roller_coasters['seating_type'].value_counts().sort_values())
plt.show()

#Average length of roller coasters per seating type (yes, length of roller coaster varies with seating type)
seating_types = roller_coasters['seating_type'].value_counts().sort_values().index
lengths = []
for type in seating_types:
    lengths.append(roller_coasters[roller_coasters['seating_type'] == type]['length'].mean())
plt.barh(y = seating_types,\
         width = lengths)
plt.show()
```

Do roller coaster manufactures have any specialties (do they focus on speed, height, seating type, or inversions)?

```python
# We check the distribution of the nb of roller coasters produced per manufacturer
plt.hist(roller_coasters['manufacturer'].value_counts(),bins=100)
plt.show()

# We're going to focus on the main manufacturers, defined as having manufactured at least 100 roller coasters in the dataset
manufacturers = []
for manufacturer in roller_coasters['manufacturer'].value_counts().index:
    if (roller_coasters['manufacturer'].value_counts()[manufacturer] >= 100) and manufacturer != 'na':
        manufacturers.append(manufacturer)

# roller_coaster_sub_main = roller_coasters[roller_coasters['manufacturer'].isin(manufacturers)].reset_index(drop = True)

# Now we draw visuals to compare the retained manufacturers according to height, seating type, nb of inversions

plt.barh(y = manufacturers,\
         width = [roller_coasters[roller_coasters['manufacturer'] == manufacturer]['height'].mean() \
                  for manufacturer in manufacturers])
plt.title('Average roller coaster height per manufacturer')
plt.show()
plt.clf()

for manufacturer in manufacturers:
    plt.barh(y=roller_coasters[roller_coasters['manufacturer'] == manufacturer]['seating_type'].value_counts().sort_values().index,\
            width =roller_coasters[roller_coasters['manufacturer'] == manufacturer]['seating_type'].value_counts().sort_values())
    plt.title(f'Seating types per manufacturer {manufacturer}')
    plt.show()
    plt.clf()


plt.barh(y = manufacturers,\
         width = [roller_coasters[roller_coasters['manufacturer'] == manufacturer]['num_inversions'].mean() \
                  for manufacturer in manufacturers])
plt.title('Average roller coaster nb of inversions per manufacturer')
plt.show()
```

Relatively to height: B&M, Pinfari, Intamin produce the highest roller coasters in average
Relatively to seating type: manufacturers all produce prominently roller coasters with 'normal' seating type, except B&M that has a very diversified range of capabilities and produces prominently roller coasters with 'inverted' seating type.
Relatively to nb. of inversions: B&M produces roller coasters with a much higher nb. of inversions in average than the competition. Apart from B&M, only Vekoma and Intamin produce roller coasters with at least 1 inversion in average.

From this analysis, B&M emerges as a player with typically more differentiated products, and so does but to a lesser extent Vekoma. Other manufacturers generally make products with 1 or even 0 inversion, standard seating type and low or average height.
